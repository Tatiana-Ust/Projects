Также одним из важнейших показателей является **Retention Rate**. Он показывает, сколько пользователей из когорты относительно их изначального числа вернулось в последующие периоды.

В данном случае когортой будут пользователи, впервые зашедшие на сайт Яндекс.Афиши в один месяц. 

Создам новый датафрейм 'first_activity_date', в котором сгруппирую одинаковые идентификаторы пользователей и найду для них самую раннюю дату начала сессии. 

Методом astype() приведу значения столбца к типу 'datetime64[M]', где [M] означает размерность временной метки - месяц. 

Полученный объект Series назову 'first_activity_month'

```python
first_activity_date = visits.groupby(['uid'])['start_ts'].min()
first_activity_month = first_activity_date.astype('datetime64[M]')
first_activity_month.name = 'first_activity_month'
first_activity_month.head()
```

Методом join() объединю датафрейм visits и Series 'first_activity_month' по столбцу 'uid':

```python
visits = visits.join(first_activity_month, on='uid')

visits.head()
```

Создам колонку 'activity_month', в которой значения столбца 'start_ts' приведу к типу 'datetime64[M]':

```python
visits['activity_month'] = visits['start_ts'].astype('datetime64[M]')
```

Рассчитаю 'lifetime' когорты: из месяца захода на сайт пользователем 'activity_date' вычту месяц когорты 'first_activity_date'.

Преобразую данные в новой колонке в число месяцев, прошедших между датами. Для этого cohort_lifetime нужно разделю на константу np.timedelta64(1, 'M') чтобы узнать, сколько величин длиной в 1 месяц содержится в cohort_lifetime.

Вследствие того, что константа np.timedelta64(1,'M') представляет собой среднюю длину месяца в году, получились дробные числа. С помощью метода round() округлю значения столбца и приведу их к целочисленному типу методом astype('int').

```python
visits['cohort_lifetime'] = (visits['activity_month'] - visits['first_activity_month'])

visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1, 'M')

visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
```

```python
visits.head()
```

Далее сгруппирую данные по когорте и 'lifetime' в датафрейме 'cohorts'. 

Для каждой когорты рассчитаю количество активных пользователей на определённый месяц 'жизни' с помощью 'nunique' в методе 'agg'

```python
cohorts = visits.groupby(['first_activity_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts.head()
```

Чтобы найти Retention Rate, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждый следующий месяц. Для этого необходимо найти исходное количество пользователей в когорте. Возьму их число на нулевом месяце.

Дополнительно переименую столбец 'uid' в 'cohort_users'

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
initial_users_count.head()
```

Объединю данные по когортам с исходным количеством пользователей в когорте:

```python
cohorts = cohorts.merge(initial_users_count,on='first_activity_month')
cohorts.head()
```

Рассчитаю Retention Rate путем деления количество активных пользователей в каждом месяце на исходное число пользователей в когорте:

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']
```

```python
cohorts.head()
```

Теперь построю сводную таблицу 'retention' и на ее основе создам тепловую карту:

```python
retention = cohorts.pivot_table(index='first_activity_month',columns='cohort_lifetime',values='retention',aggfunc='sum')
retention
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(15, 10))
plt.title('Cohorts: User Retention')
retention.index = [str(x)[0:10] for x in retention.index]
sns.heatmap(retention, annot=True, fmt='.1%', linewidths=1, linecolor='grey', vmax=0.1); 
```

Посмотрю средний Retention Rate на второй месяц 'жизни' когорт:

```python
round((cohorts[cohorts['cohort_lifetime'] == 2][['retention']].mean()*100),2)
```

В данном разделе будут проанализированы такие показатели, как:
- среднее количество времени от первого посещения сайта до покупки;
- среднее количество покупок на одного покупателя;
- средний чек;
- LTV - сумма денег, которую клиент приносит в среднем со всех своих покупок.

Для исследования **времени между входом на сайт и покупкой**, методом join() объединю данные о заказах ('orders') и данные о первой сессии пользователя на сайте Series 'first_activity_date'. 

Чтобы избежать путаницы, переименую новый столбец в 'first_ts'

Далее рассчитаю **количество покупок на одного покупателя, в том числе среднее значение за полгода** применяя когортный анализ.

Для этих целей использую датафрейм orders. 

Сначала сгруппирую данные по уникальному идентификатору пользователя и найду для каждого самую раннюю дату заказа:

```python
first_order_month = orders.groupby(['uid'])['buy_ts'].min()
first_order_month = first_order_month.astype('datetime64[M]')
first_order_month.name = 'first_order_month'
first_order_month.head()
```

```python
orders['buy_month'] = orders['buy_ts'].astype('datetime64[M]')
```

В колонке 'cohort_lifetime' будет содержаться разница между месяцем очередного заказа и месяцем первого заказа. Значения данного столбца будут являться 'временем жизни' каждой когорты

```python
orders['cohort_lifetime'] = (orders['buy_month'] - orders['first_order_month'])
orders['cohort_lifetime'] = orders['cohort_lifetime'] / np.timedelta64(1, 'M')
orders['cohort_lifetime'] = orders['cohort_lifetime'].round().astype('int')
orders.head()
```

Теперь создам датафрейм 'orders_per_user', в котором сгруппирую данные по когорте и 'времени жизни' каждой когорты. 

Также рассчитаю количество заказов пользователей по каждой когорте в каждом ее 'времени жизни' и количество пользователей, сделавших заказ. Назову колонки 'amount_orders' и 'amount_users', соответственно.

В столбце 'orders_per_user' будут содержаться значения о количестве заказов на 1 пользователя из каждой когорты.

```python
orders_per_user = orders.groupby(['first_order_month','cohort_lifetime']).agg({'uid': ['count','nunique']})
orders_per_user.columns = ['amount_orders', 'amount_users']
orders_per_user['orders_per_user'] = orders_per_user['amount_orders'] / orders_per_user['amount_users']
orders_per_user = orders_per_user.reset_index()
orders_per_user.head()
```

С помощью метода pivot_table создам датафрейм 'purchases_amount', в котором будут отображены все когорты в разрезе каждого месяца их 'жизни' и количество заказов на 1 пользователя:

```python
purchases_amount = orders_per_user.pivot_table(index='first_order_month',
                                               values='orders_per_user',
                                               columns='cohort_lifetime',
                                               aggfunc='sum')
purchases_amount = round(purchases_amount,2)
```

```python
purchases_amount
```

Теперь посмотрю среднее значение количества заказов за полгода:

```python
round((purchases_amount[5].mean()),2)
```

**Средний чек** - это товарооборот в денежном выражении, приведённый к количеству заказов за определенный период, 

Средний чек, не смотря на относительную простоту рассчета, является комплексным показателем, демонстрирующим насколько продуман ассортимент товаров и услуг, насколько грамотен мерчандайзинг и маркетинговая стратегия. Рассчитывается как сумма всех совершенных клиентами покупок за определенный период времени, деленная на количество заказов за тот же период.

Для начала создам датафрейм 'avg_oder', в котором сгруппирую данные по времени заказа:

```python
avg_oder = orders.groupby('buy_month').agg({'revenue': 'sum', 'uid': 'nunique'})
```

```python
avg_oder.head()
```

Теперь найду средний чек покупателя в колонке 'revenue_per_user', путем деления суммы заказа на количество заказов:

```python
avg_oder['revenue_per_user'] = avg_oder['revenue']/avg_oder['uid']
```

Построю сводную таблицу изменения среднего чека в по месяцу совершения покупки и посмотрю, как изменяется средний чек с течением времени:

```python
avg_oder.pivot_table(index='buy_month',
                     values='revenue_per_user',
                     aggfunc='mean').plot(figsize = (15, 5),
                                          grid = True,
                                          cmap = 'Pastel2')
plt.xlabel('Месяц заказа')
plt.title('Изменения среднего чека во времени')
plt.legend(['Средний чек']);
```

**LTV**  — «пожизненная» ценность клиента. Это — итоговая сумма денег, которую один клиент приносит в среднем со всех своих покупок. При расчёте LTV обычно обращаются к валовой прибыли с клиента за период.

Посчитаю количество новых покупателей 'new_buyers' за каждый месяц. Для этого создам датафрейм 'new_buyers' путем группировки датафрейма 'orders' по месяцу первого заказа

```python
new_buyers = (orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index())
new_buyers.columns = ['first_order_month', 'new_buyers']
new_buyers.head()
```

Далее сгруппирую данные о заказах по месяцу первой покупки и месяцу каждого заказа в датафрейме 'cohorts_rev'. Также просуммирую выручку

```python
cohorts_rev = orders.groupby(['first_order_month', 'buy_month']).agg({'revenue': 'sum'}).reset_index()
print(cohorts_rev.head())
```

Так как с точки зрения когортного анализа LTV — накопительная метрика, добавлю в таблицу 'cohorts_rev' данные о том, сколько людей первый раз совершили покупку в каждый месяц путем создания датафрейма 'total_orders':

```python
total_orders = pd.merge(new_buyers, cohorts_rev, on='first_order_month')
print(total_orders.head())
```

Поскольку LTV считают из валовой прибыли, значит, нужно ее найти, умножив выручку на маржинальность. 

Так как LTV является относительным показателем, его удобнее смотреть по 'времени жизни' когорты.

Маржинальность сервиса — 100%, поэтому задам показатель margin_rate = 1.

```python
margin_rate = 1

total_orders['proceeds'] = total_orders['revenue'] * margin_rate
total_orders['cohort_lifetime'] = (total_orders['buy_month'] - total_orders['first_order_month'])/np.timedelta64(1, 'M')
total_orders['cohort_lifetime'] = total_orders['cohort_lifetime'].round().astype('int')
print(total_orders.head())
```

Настало время найти непосредственно LTV. Для этого разделю валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте. 

Результат выведу в виде сводной таблицы 'ltv':

```python
total_orders['ltv'] = total_orders['proceeds'] / total_orders['new_buyers']

ltv = total_orders.pivot_table(index='first_order_month', 
                               columns='cohort_lifetime', 
                               values='ltv', 
                               aggfunc='mean')
ltv = round(ltv.cumsum(axis=1),2)
ltv.fillna('')
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(15, 10))
plt.title('Cohorts: LTV')
ltv.index = [str(x)[0:10] for x in ltv.index]
sns.heatmap(ltv, annot=True, linewidths=1, linecolor='grey', cmap= 'coolwarm'); 
```

```python
round(ltv[5].mean(),2)
```
