Так как пропущенных значений в колонке 'locality_name' даже меньше половины процента, нет смысла делать замену пропущенных значений. 
Таким образом, целесообразнее исключить из обработки строки с пропущенными значениями в столбце 'locality_name' для упрощения дальнейшего анализа.

```python
data = data.dropna(subset = ['locality_name']).reset_index(drop = True)
```

```python
print('Количество пропусков в колонке "locality_name":', data['locality_name'].isnull().sum())
```

Аналогично предыдущей ситуации с пропусками названий населенных пунктов, пропуски количества этажей составляют менее 1% от общего количества.
    
Можно было бы заменить пропущенные значения на соответствующее индексу значение этажа, на котором находится квартира (по колонке 'floor'). Однако нет гарантии, что квартира находится на последнем этаже. При этом, если в последствии делать выборку квартир, находящихся именно на последнем этаже, искусственно добавленные на этом этапе квартиры могут дать значительную погрешность.

Принимая во внимание вышеописанное, наиболее логичным будет оставить строки с пропусками в колонке 'floors_total' без изменения.

Заменю пропущенные значения по колонке **'balcony'** на 0 с помощью fillna(). Этим предположу, что если их количество не указано, значит в квартире нет балконов

```python
data['balcony'] = data['balcony'].fillna(0)
```

Проверю количество пропусков после замены:

```python
print('Количество пропусков в колонке "balcony":', data['balcony'].isnull().sum())
```

Больше всего пропущенных значений колонке **'is_apartment'**. Можно предположить, что такая ситуация возникла по причине незаполнения значениями этой колонки в случае, если квартира не является аппартаментами. Также возможно что при конвертации данных нулевые значения, т.е. False превратились в пропуски, т.е. NaN.
В результате из-за пропущенных значений колонка 'is_apartment' имеет тип 'object', а не 'bool'.
Исправлю эту проблему с помощью метода fillna() и заменю пропуски на False:

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Проверю тип данных в вышеуказанной колонке:

```python
data['is_apartment'].dtypes
```

```python
print('Количество пропусков в колонке "is_apartment":', data['is_apartment'].isnull().sum())
```

В результате столбец 'is_apartment' больше не содержит пропусков и принадлежит к булеву типу.

Замену пропусков в колонке **'airports_nearest'** (расстояние до ближайшего аэропорта в метрах) осуществлю с помощью метода transform в зависимости от названия населенного пункта и медианного значения:

```python
data['airports_nearest'] = data['airports_nearest'].fillna(data.groupby('locality_name')['airports_nearest'].transform('median'))
```

```python
print('Количество пропусков в колонке "airports_nearest":', data['airports_nearest'].isnull().sum())
```

Количество пропусков сократилось, но не исчезло полностью, значит замена сработала, но не до конца. Можно предположить, что проблема в названиях населенных пунктов

```python
print('В столбце "locality_name"', data['locality_name'].nunique(), 'уникальных значения')
```

```python
data['locality_name'].unique()
```

При просмотре названий населенных пунктов становится заметно, что некоторые названия записаны по разному, например, в словосочетаниях со словами "поселок" или "деревня"

С помощью метода replace с указанием в параметре regex = True исключу все лишние значения в названиях населенных пунктов:

```python
data['locality_name'] = data['locality_name'].replace('пос[её]лок городского типа |пос[её]лок |деревня |городского типа |городской |село |садоводческое некоммерческое товарищество |коттеджный |садовое товарищество |станции |при железнодорожной станции |при железнодорожной |имени ', '', regex = True)
```

```python
print('После обработки осталось', data['locality_name'].nunique(), 'уникальных значений')
```

```python
data['airports_nearest'] = data['airports_nearest'].fillna(data.groupby('locality_name')['airports_nearest'].transform('median'))
```

```python
print('Количество пропусков в колонке "airports_nearest":', data['airports_nearest'].isnull().sum())
```

```python
data['locality_name'].unique().sort()
```

Таким образом, если количество пропусков осталось неизменным после исключения "лишних" названий населенных пунктов, значит невозможно найти соответствующее медианное значение по названию населенного пункта. 
Так как указанный столбец не участвует в дальнейшем анализе, заменять пустые значения не имеет смысла

Пропуски в колонке **'cityCenters_nearest'** (расстояние до центра города (м)) также осуществлю с помощью метода transform в зависимости от названия населенного пункта и медианного значения:

```python
data['cityCenters_nearest'] = data['cityCenters_nearest'].fillna(data.groupby('locality_name')['cityCenters_nearest'].transform('median'))
```

```python
print('Количество пропусков в колонке "cityCenters_nearest":', data['cityCenters_nearest'].isnull().sum())
```

В целом, можно было найти в таблице названия населенных пунктов с неизвестными расстояниями до центра города или аэропорта и подставить реальные значения в пропуски, но это отнимет очень много времени.

Пропуски в колонках **'parks_around3000'** (число парков в радиусе 3 км), **'parks_nearest'** (расстояние до ближайшего парка (м)), **'ponds_around3000'** (число водоемов в радиусе 3 км), **'ponds_nearest'** (расстояние до ближайшего водоема (м)), также осуществлю с помощью метода transform в зависимости от названия населенного пункта и медианного значения:

## Расчёты и добавление результатов в таблицу

### Стоимость квадратного метра

Стоимость квадратного метра рассчитаю как результат деления цены квартиры на общую площадь. Для этого создам новый столбец 'square_meter. Сразу поменяю его тип на int и выведу первые 10 строк нового столбца и столбцов, используемых для рассчетов:

```python
data['square_meter'] = (data['last_price']/data['total_area'])
data['square_meter'] = data['square_meter'].astype('int')
print(data.loc[:, ['square_meter', 'last_price', 'total_area']].head(10))
```

### День недели, месяц и год публикации объявления

С помощью методов dt.dayofweek, dt.month и dt.year найду значения номера дня недели, порядковый номера месяца и год первой публикации объявления. Полученные данные занесу в колонки 'day_exposition', 'month_exposition' и 'year_exposition' - соответственно.

```python
data['day_exposition'] = data['first_day_exposition'].dt.dayofweek
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
print((data.loc[:, ['day_exposition', 'month_exposition', 'year_exposition']]).head())
```

В колонке 'day_exposition' заменю чистовые значения на названия дней недели:

```python
def days(row):
    if row['day_exposition'] == 0:
        return 'Воскресенье'
    if row['day_exposition'] == 1:
        return 'Понедельник'
    if row['day_exposition'] == 2:
        return 'Вторник'
    if row['day_exposition'] == 3:
        return 'Среда'
    if row['day_exposition'] == 4:
        return 'Четверг'
    if row['day_exposition'] == 5:
        return 'Пятница'
    if row['day_exposition'] == 6:
        return 'Суббота'
    return row


data['day_exposition'] = data.apply(days, axis=1)
```

### Этаж квартиры

С помощью функции 'floors' категоризирую этажи на 3 типа:первый этаж, последний этаж и другой этаж. Получившиеся значения добавлю в новую колонку 'apartment_floor'

```python
def floors(row):
    if row['floor'] == row['floors_total']:
        return 'Последний этаж'
    if row['floor'] == 1:
        return 'Первый этаж'
    return 'Другой этаж'


data['apartment_floor'] = data.apply(floors, axis=1)
print((data.loc[:, ['floor', 'floors_total', 'apartment_floor']]).head())
```

**Вывод**

В исходные данные были добавлены дополнительные колонки:
* 'square_meter' - с информацией о стоимости квадратного метра квартиры;
* 'first_day_exposition', 'first_day_exposition', 'first_day_exposition' - значения номера дня недели, порядкового номера месяца и года первой публикации объявления;
*  'apartment_floor' - содержит информацию на каком этаже располагается анализируемая квартира: на первом, последнем или другом.

#### Зависимость стоимости квартиры от площади

Для более корректного результата анализа разобью колонку 'total_area' на квантили, составлю словарь 'dict_total_area'. Далее поделю все значения по категорям и добавлю соответствующую колонку в data.

```python
dict_total_area = data['total_area'].quantile([.25,.5,.75]).to_dict()
print(dict_total_area)

def area(row):
    if row['total_area'] <= dict_total_area[0.25]:
        return 'Небольшая общая площадь';
    if row['total_area'] <= dict_total_area[0.5]:
        return 'Средная общая площадь';
    if row['total_area'] <= dict_total_area[0.75]:
        return 'Большая общая площадь';
    return 'Очень большая общая площадь'


data['dict_total_area'] = data.apply(area, axis=1)
```
